### Section 1: The frequentist interpretation of probability

There are two main philosphies present in probability and statistics: the frequentists interpretation and the Bayesian interpretation. For the majority of this course, the two viewpoints will agree, but it may be easier to refer to one viewpoint over the other when explaining certain concepts. 

The frequentist interpretation of probability is essentially that the probability of an event is the limit of its relative frequency after an infinite number of trials. The idea put into practice is that the probability of an event is fixed, and as we conduct more and more experiments, the frequency of outcomes will converge on the true probabilities of the outcomes. We can see this in practice with simple experiments like flipping a fair coin. After only a few flips, we may have achieved a very uneven number of heads and tails, but as we flip the coin more and more times, the frequency of heads will begin to approach the frequency of tails. 

The Bayesian interpretation of probability is that the probability of an event is determined by the previous evidence about the event and the observers expectations of an outcome. The probability of an event can therefore be updated as we run more and more experiments and get more information. In the simple coin flipping experiment, the interpretation would be that we are not sure if the coin is fair or not. After a small number of flips, we would expect that the next flip would fall in line with our previous observations. If the coin <i>is</i> fair, then after a very large number of flips we would see the probability of tails still approaching the probability of heads. 

For most of this course, we will find the frequentist interpretation to be the easier interpretation to adopt.

### Section 2: Generating random numbers

For this lab, we will be exploring the frequentist interpretation of probability by simulating some simple experiments. In order to simulate something like flipping a coin, we will need to be able to incorporate randomness somehow. The way we will do this is through a random number generator (RNG). The random number generators we use in python do not generate numbers <i>truly</i> randomly, but by what is called a pseudo-random process. Pseudorandom processes are deterministic processes that look random, but can be reproduced perfectly if we use the same starting point (called a seed). The most commonly used pseudorandom number generator is called the Mersenne Twister.

There are several ways to generate random numbers in python, but the way we will focus on is using the `numpy` package. To get started we'll import numpy as np. The numpy module has a submodule called `random`, wich we can use to get random numbers. The following code block imports numpy, seeds the random number generator at 1, and then generates a random number between 0 and 1. Since we've seeded the rng, the outcome of `np.random.rand()` is 0.417022004702574. If we change the seed, or don't seed the rng, we will get a different outcome. 

```python
import numpy as np #Import numpy
np.random.seed(1) #Seed the pseudo-random number generator at 1
np.random.rand() #Generate a random number between 0 and 1
```

### Section 3: Plotting

Plotting can be done in several ways using python. We will use the library `pyplot`. To get started, we'll import pyplot as plt. Plotting can be done using the plot function in the pyplot package if we want to connect the points with line segments, or the scatter function if we just want to plot the points. These functions take two arguments, the first is a list of x-values, the second is a list of y-values. The pyplot package has many functions for altering the appearance of the plot after we have created the plot. We are not typically interested in being too creative with the plots, although you may want to explore a little bit the [pyplot documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.html) or by looking online for a pyplot tutorial. 

```python
import matplotlib.pyplot as plt #Import Pyplot
x = np.array(range(1,10)) #Create an array of the numbers from 1 to 9
y = np.random.rand(9) #Create an array of 9 random numbers between 0 and 1
plt.scatter(x,y) #Plot the x-y pairs as a scatter plot
plt.xlabel("Numbers from 1 to 9") #Change the x label
plt.ylabel("Random Numbers") #Change the y label
plt.title("My pretty plot") #Change the title
```
