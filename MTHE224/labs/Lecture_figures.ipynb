```python
import numpy as np
from matplotlib import pyplot as plt

t_range = np.linspace(0,10,25)
y_range = np.linspace(0,2,25)
t,y = np.meshgrid(t_range,y_range)

def F(t,y):
    return 2*y*(1-y)

vt = F(t,y)
ut = np.ones_like(vt)
v = vt / np.sqrt(ut**2+vt**2)
u = ut / np.sqrt(ut**2+vt**2)

plt.quiver(t,y,u,v)
```

```python
def solution(t,C):
    return C*np.exp(t)/(1+C*np.exp(t))
plt.style.use('dark_background')
t_vals = np.linspace(0,10,500)
y_1 = [solution(t,1) for t in t_vals]
y_3 = [solution(t,-3) for t in t_vals]
y_4 = [solution(t,0.1) for t in t_vals]
plt.quiver(t,y,u,v,color="w")
plt.plot(t_vals,y_1)
plt.plot(t_vals,y_4)
plt.plot(t_vals,y_3)

plt.savefig('../lectures/images/logistic.png', transparent = True)
```
